**1. What is the difference between a function and a method in Python?**

In Python, functions and methods are similar in that they both represent callable pieces of code, but they differ in how they are used and associated with objects. Here's a detailed explanation:

1.Function:

.A function is a standalone block of code that is defined using the def keyword.

.It can take input arguments and may return a value.

.Functions can be called independently, without being associated with an object.

```python
def greet(name):
    return f"Hello, {name}"

# Function call
greet("Alice")
```

2.Method:

.A method is a function that is associated with an object. It is defined within a class and operates on instances of that class.

.Methods are called on objects and have access to the data (attributes) of that object.

.The first parameter of a method is usually self, which refers to the instance of the class it is being called on.

```python
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, {self.name}"

# Create an object of the class
person = Person("Alice")

# Method call
person.greet()
```

**6. Explain the concept of generators in Python and how they are defined.**

In Python, generators are a type of iterable, like lists or tuples, but they allow you to iterate through a sequence of values lazily—meaning they generate items one at a time, only when needed, rather than storing everything in memory at once. This makes them memory efficient, especially when working with large datasets or infinite sequences.

1.Key Characteristics of Generators:

.Memory Efficiency: Generators do not store the entire sequence in memory. Instead, they generate values on-the-fly, which makes them more memory efficient for large data or infinite sequences.

.Lazy Evaluation: Generators yield items one at a time as they are needed, making them more suitable for large or indefinite data streams.

.Single Iteration: Generators can only be iterated over once. After all items are yielded, they are "exhausted" and cannot be reused unless regenerated.

2.How Generators are Defined:

Generators can be created in two main ways:

.Using Generator Functions (defined with the yield keyword).

.Using Generator Expressions (similar to list comprehensions, but with parentheses).

1. Generator Functions:

A generator function is defined like a normal function using def, but instead of using return to send back a result, it uses the yield keyword. Each time the yield statement is encountered, the function "yields" a value and pauses its state. When the function is called again, it resumes from where it left off.

```python
def count_up_to(max_value):
    count = 1
    while count <= max_value:
        yield count
        count += 1

# Creating a generator
counter = count_up_to(5)

# Iterating over the generator
for num in counter:
    print(num)
```

```python
#How to Use next() with Generators:

#You can also manually retrieve values from a generator using the next() function:


counter = count_up_to(3)

print(next(counter))
print(next(counter))
print(next(counter))
```

```python
2. Generator Expressions:
A generator expression is a concise way to create generators, similar to how list comprehensions work, but with parentheses instead of square brackets. This is useful for creating simple, memory-efficient generators.
```

```python
# Generator expression to create squares of numbers
squares = (x ** 2 for x in range(5))

# Iterating over the generator
for square in squares:
    print(square)
```

Why Use Generators?

.Memory Efficiency: For large datasets or streams of data, storing everything in memory can be inefficient or impossible. Generators solve this problem by producing values only when required.

.Handling Infinite Sequences: Generators are ideal for working with infinite sequences, like generating Fibonacci numbers, because they don’t need to hold all values in memory at once.

```python
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

# Creating an infinite Fibonacci generator
fib = fibonacci()

# Getting the first 10 Fibonacci numbers
for _ in range(10):
    print(next(fib))
```

**4.Write a Python function that checks if a given number is prime or not from 1 to 200**

To check if a given number is prime, you can create a Python function that verifies the conditions of prime numbers (a prime number is greater than 1 and has no divisors other than 1 and itself). Here’s how you can implement such a function and check for prime numbers between 1 and 200:

```python
def is_prime(n):
    if n <= 1:
        return False  # Numbers less than 2 are not prime
    for i in range(2, int(n**0.5) + 1):  # Check for factors from 2 to sqrt(n)
        if n % i == 0:
            return False  # Found a divisor, not prime
    return True  # No divisors found, it is prime

# Example usage: Check prime numbers from 1 to 200
prime_numbers = [num for num in range(1, 201) if is_prime(num)]
print(prime_numbers)
```

**5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.**

You can create an iterator class in Python to generate the Fibonacci sequence up to a specified number of terms. Here’s an implementation of such a class:



```python
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.current = 0
        self.next_value = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.terms:
            # Generate the next Fibonacci number
            fib_number = self.current
            self.current, self.next_value = self.next_value, self.current + self.next_value
            self.count += 1
            return fib_number
        else:
            raise StopIteration  # No more terms to generate

# Example usage
fibonacci_sequence = FibonacciIterator(10)  # Create an iterator for the first 10 Fibonacci numbers
for num in fibonacci_sequence:
    print(num)
```

```python
Explanation:

1.Class Initialization:


The FibonacciIterator class is initialized with the number of terms (terms) to generate.

current starts at 0, next_value starts at 1, and count tracks how many terms have been generated.


2.Iterator Protocol:

The __iter__ method returns the iterator object itself.

The __next__ method generates the next Fibonacci number:

    If the current count is less than the specified number of terms, it calculates the next Fibonacci number and updates the values of current and next_value.

    If the specified number of terms has been reached, it raises a StopIteration exception to signal that the iteration is complete.


3.Example Usage:

An instance of FibonacciIterator is created for the first 10 Fibonacci numbers.

A for loop is used to iterate over the generated Fibonacci numbers and print them.
```

```python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()  # Yield each line without trailing newline characters

# Example usage
file_path = 'example.txt'  # Replace with your file path
for line in read_file_line_by_line(file_path):
    print(line)
```

```python
Explanation:

Function Definition: The read_file_line_by_line function takes a parameter file_path, which is the path to the file to be read.

File Handling:


The function uses a with statement to open the file. This ensures that the file is properly closed after reading, even if an error occurs.

The file is opened in read mode ('r').

Reading Lines:

A for loop iterates over each line in the file.

The yield statement is used to return each line after stripping any leading or trailing whitespace (including newline characters) with line.strip().

Example Usage:

Replace 'example.txt' with the path to your file.

The generator is called in a for loop, which prints each line from the file
```

**8.Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.**

You can use a lambda function in Python to sort a list of tuples based on the second element of each tuple. The built-in sorted() function can take a key argument, where you can specify a lambda function that extracts the second element. Here’s how you can do that:

```python
# Sample list of tuples
data = [(1, 'apple'), (3, 'banana'), (2, 'cherry'), (4, 'date')]

# Sorting the list of tuples based on the second element
sorted_data = sorted(data, key=lambda x: x[1])

# Example output
print(sorted_data)
```

```python
Explanation:

List of Tuples: The data variable contains a list of tuples. Each tuple consists of an integer and a string.

Sorting with sorted():

The sorted() function is called with the data list as the first argument.

The key parameter is set to a lambda function lambda x: x[1], which extracts the second element of each tuple (x[1]).

Output: The result of the sorting operation is stored in sorted_data, which will contain the tuples sorted based on the second element (the string). The output will be:
```

**9. Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit**

You can use the map() function in Python to convert a list of temperatures from Celsius to Fahrenheit. The formula for converting Celsius to Fahrenheit is:

```python
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temperatures = [0, 20, 37, 100]

# Using map() to convert Celsius to Fahrenheit
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Example output
print(fahrenheit_temperatures)
```

```python
Explanation:
Conversion Function: The celsius_to_fahrenheit function takes a temperature in Celsius and returns the corresponding temperature in Fahrenheit.

List of Temperatures: The variable celsius_temperatures contains a list of temperatures in Celsius.

Using map():

      .The map() function applies the celsius_to_fahrenheit function to each temperature in the celsius_temperatures list.

      .The result of map() is converted to a list using list().

Output: Finally, the converted temperatures in Fahrenheit are printed.
```

**10.Create a Python program that uses filter() to remove all the vowels from a given string**


You can use the filter() function in Python to remove all the vowels from a given string. The filter() function takes a function and an iterable as arguments and returns an iterator containing only the elements for which the function returns True. Here's how you can implement this:

```python
def remove_vowels(input_string):
    vowels = "aeiouAEIOU"  # Define the vowels to be removed
    # Use filter to remove vowels from the input string
    filtered_string = ''.join(filter(lambda x: x not in vowels, input_string))
    return filtered_string

# Example usage
input_string = "Hello, World!"
result = remove_vowels(input_string)
print(result)
```

```python
Explanation:

Function Definition: The remove_vowels function takes a string input_string as its parameter.

Vowels Definition: The variable vowels contains all the vowels (both lowercase and uppercase) that we want to remove from the string.

Using filter():

The filter() function is used with a lambda function that checks if each character x in input_string is not in the vowels string.

The result of filter() is an iterator, so we use ''.join() to concatenate the remaining characters into a new string.

Example Usage:

The function is called with the string "Hello, World!", and the result is printed.
```

**Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,- € if the value of the order is smaller than 100,00 €.**


Here’s a Python program that processes the given data (order number, book title and author, quantity, price per item) and returns a list of 2-tuples. Each tuple contains the order number and the total price for the order, with an additional €10 charge if the order's total is less than €100.

```python
def calculate_order_totals(orders):
    result = []

    for order in orders:
        order_number, _, quantity, price_per_item = order
        total_value = price_per_item * quantity

        # Add 10 if total value is less than 100
        if total_value < 100:
            total_value += 10

        result.append((order_number, total_value))

    return result

# Orders: (order_number, book_title_and_author, quantity, price_per_item)
orders = [
    (34587, "learning python, mark lutz", 4, 40.95),
    (98762, "programming python, mark lutz", 5, 56.80),
    (77226, "head first python, paul barry", 3, 32.95),
    (88112, "einfuhrung in python3, bernd klein", 3, 24.99)
]

# Calculate the order totals
order_totals = calculate_order_totals(orders)

# Output the result
print(order_totals)
```
